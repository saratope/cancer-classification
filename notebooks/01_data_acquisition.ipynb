```python
import GEOparse
import pandas as pd
```

```python
#Get GEO
gse = GEOparse.get_GEO("GSE45827", destdir="../data/", silent=False)
```

```python
#Get metadata
sample_meta = pd.DataFrame.from_dict(
    {gsm: gse.gsms[gsm].metadata for gsm in gse.gsms},
    orient="index"
)
sample_meta
```

```python
# Extract expression matrix
data_matrix = gse.pivot_samples("VALUE")
data_matrix.to_csv("../data/raw_expression_matrix.csv")
```
